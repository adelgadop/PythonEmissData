```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
voc_frac = pd.read_csv("../2_Emissions_inventory/EDGAR-HTAP/voc_frac.csv")
voc_frac.set_index('voc', inplace=True)
voc_frac
```

```python
fig, ax = plt.subplots(figsize=(6,2))
voc_frac.plot(kind='bar',y='per', ax=ax, edgecolor='k',
              ylabel='Percentage [%]', rot=90, legend=False)
ax.set_xlabel('VOC species')
fig.savefig("../4_Draft_Report/Dissertation/fig/voc_frac.pdf",
           bbox_inches='tight', facecolor='w')
```
